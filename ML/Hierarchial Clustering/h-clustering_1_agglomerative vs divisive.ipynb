```python
# Agglomerative vs Divisive Clustering

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering, BisectingKMeans
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
# ---
# 1) Create simple dataset
# ---
X, _ = make_blobs(
    n_samples=200,
    centers=3,
    cluster_std=0.9,
    random_state=42
)
```

```python
# 2) AGGLOMERATIVE (Bottom-up)
# 
agg = AgglomerativeClustering(n_clusters=3, linkage="ward")  # minimizes the sum of squared differences
agg_labels = agg.fit_predict(X)

# 3) DIVISIVE (Top-down)
# (Bisecting k-Means is a practical divisive method)
# 
div = BisectingKMeans(n_clusters=3, random_state=42)
div_labels = div.fit_predict(X)
div_labels
```

```python
# 4) Visual comparison
fig, axes = plt.subplots(1, 2, figsize=(9,4))
axes[0].scatter(X[:,0], X[:,1], c=agg_labels, cmap="tab10", s=15)
axes[0].set_title("Agglomerative (Bottom-Up)")
axes[1].scatter(X[:,0], X[:,1], c=div_labels, cmap="tab10", s=15)
axes[1].set_title("Divisive (Top-Down)")
plt.tight_layout()
plt.show()
```
