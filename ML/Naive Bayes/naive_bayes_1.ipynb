```python
# Demonstration of Naive Bayes and the feature independence assumption

import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# ---
# 1) Load dataset
# ---
data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Number of features:", X_train.shape[1])

# ---
# 2) Train Naive Bayes classifier
# ---
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
# 3) Evaluate performance
# ---
y_pred = nb.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f"Naive Bayes Accuracy: {acc:.4f}")

# ---
# 4) Show Learned probabilities (for intuitive)
# ---
print("\nClass prior probabilities P(y):")
print(nb.class_prior_)

print("\nFeature means per class (sample of first 5 features):")
print(nb.theta_[:, :5])

print("\nFeature variances per class (sample of first 5 features):")
print(nb.var_[:, :5])
```
