```python
# 1. Import Required Libraries
# ===========================================
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 2. Load Dataset
# ===========================================
# Breast Cancer dataset is a real-world medical dataset
data = load_breast_cancer()

# Features and target
X = data.data
y = data.target

# Convert to DataFrame for better understanding
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

print("Dataset Shape:", df.shape)
print("\nFirst 5 rows of dataset:\n")
print(df.head())
```

```python
# 3. Dataset Information
# ======================
print("\nTarget Classes:")
print(data.target_names)

print("\nClass Distribution:")
print(df['target'].value_counts())

# ======================
# 4. Split Dataset into Training and Testing Sets
# ======================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("\nTraining samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# 5. Create and Train Naive Bayes Model
# ==========================================
# GaussianNB is used for continuous numerical data
nb_model = GaussianNB()

# Train the model
nb_model.fit(X_train, y_train)

print("\nNaive Bayes model training completed.")

# ==========================================
# 6. Make Predictions
# ==========================================
y_pred = nb_model.predict(X_test)

print("\nFirst 10 Predictions:")
print(y_pred[:10])
```

```python
# 7. Model Evaluation
# ======================

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", round(accuracy * 100, 2), "%")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

print("\nConfusion Matrix:")
print(cm)

# Detailed classification metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# ====================
# 8. Interpretation of Confusion Matrix
# ====================
print("\nConfusion Matrix Interpretation:")
print("True Negative (TN):", cm[0][0])
print("False Positive (FP):", cm[0][1])
print("False Negative (FN):", cm[1][0])
print("True Positive (TP):", cm[1][1])
```

```python
# ====================================
# 9. Predict on New Sample (Example)
# ====================================
sample = X_test[0].reshape(1, -1)
prediction = nb_model.predict(sample)

print("\nPrediction for a new sample:",
    data.target_names[prediction][0])
```
