```python
import matplotlib.pyplot as plt
days=[1, 2, 3, 4, 5,6,7]
sales=[100, 120, 150, 130, 170, 160, 180]
plt.plot(days,sales,marker='o',linewidth=2,alpha=0.8,label='Daily Sales')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Daily Sales Over a Week')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("ecommerce_sales_analytics_5000.csv",parse_dates=['order_date'])
df=df.sample(n=100)
```

```python
monthly_revenue=(df.set_index('order_date')['revenue'].resample('ME').sum())
fig, ax=plt.subplots(figsize=(7,4))
ax.plot(monthly_revenue.index,monthly_revenue.values,linewidth=2)
ax.set_xlabel('Month')
ax.set_ylabel('Total Revenue')
ax.set_title('Monthly Revenue Trend')
```

```python
fig, ax=plt.subplots(figsize=(7,4))
ax.scatter(df['unit_price'],df['revenue'],alpha=0.6)
ax.set_xlabel('Unit Price')
ax.set_ylabel('Revenue')
ax.set_title('Sales Analysis')
```

```python
x=np.linspace(0, 10, 100)
y=np.sin(x)
fig, axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(x,y,color='b')
axs[0,0].set_title('Line Chart')
axs[0,1].scatter(x,y,color='r')
axs[0,1].set_title('Scatter Plot')
axs[1,0].hist(np.random.randn(1000),bins=20,color='g',alpha=0.7)
axs[1,0].set_title('Histogram')
categories=['A','B','C','D']
values=[10,15,8,12]
axs[1,1].bar(categories,values,color='orange')
axs[1,1].set_title('Bar Chart')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
df=pd.read_csv("healthcare_patient_analytics_seaborn.csv")
```

```python
df =df.sample(500)
sns.scatterplot(data=df,x="treatment_cost",y="recovery_score",hue="visit_type")
plt.title("Treatment Cost vs Recovery Score")
plt.show()
```

```python
sns.scatterplot(data=df,x="treatment_cost",y="recovery_score")
```

```python
sns.kdeplot(data=df,x="recovery_score",hue="visit_type",fill=True,)
plt.title("KDE Recovery Score Distribution by Visit Type")
plt.show()
```

```python
 sns.ecdfplot(data=df,x="length_of_stay_days",hue="visit_type")
 plt.title("ECDF Length of Stay Distribution by Visit Type")
 plt.show()
```

```python
df=pd.DataFrame({"City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
                 "Segment": ["A", "B", "A", "B", "A"],
                    "Revenue": [1000, 1500, 1200, 1300, 1100]})
print(df)
```

```python
encoded=pd.get_dummies(df,columns=["City","Segment"])
print(encoded)
```

```python
from sklearn.preprocessing import OrdinalEncoder
df2=pd.DataFrame({"Size": ["Small", "Medium", "Large", "Medium", "Small"]})
encoder=OrdinalEncoder(categories=[["Small","Medium","Large"]])
df2["Size_Encoded"]=encoder.fit_transform(df2[["Size"]])
print(df2)
```
